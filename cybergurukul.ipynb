````python
## this is for the nTH fibonacci no
def fibonacci(num):
    if num==0:
        return 0
    if num==1:
        return 1
    return fibonacci(num-1) + fibonacci(num-2)

fibonacci(7)

## list of the Nth fibonacci series

def fibonacci_series(num):
    if num==0:
        return 0
    if num==1:
        return [0, 1]
    result=[0,1]
    for i in range(2,num):
        result.append(result[i-1]+result[i-2])  #3  1+1=2
    return result

fibonacci_series(7)

````

````python
def findingmissing(arr):
    length=len(arr)+1
    totalsum=length*(length+1)//2

    sumarr=sum(arr)
    return totalsum-sumarr
findingmissing([1,2,3,4,5,6,8,9])
````

**Test Cases:**
```python

```

---

````python
def removeduplicate(string1):
    seen=set()
    result=""
    for i in string1:
        if i not in seen:
            result+=i
            seen.add(i)
    return result

removeduplicate("banana")
````

````python
def are_anagrams(s1,s2):
    if len(s1)!=len(s2):
        return False
    else:
        s1=sorted(s1)
        s2=sorted(s2)
        if s1==s2:
            return True
        else:
            return False
are_anagrams("aabb","baba")
````

````python
def sumdigits(num):
    if not num:
        return "enter valid input"
    numstr=str(num)
    sumnum=0
    for i in numstr:
        sumnum+=int(i)

    return sumnum

sumdigits(123)
````

````python
import io
import contextlib

# Automated scoring for notebook programs

def _get_fn(name):
    return globals().get(name)

def _capture_stdout(fn, *args, **kwargs):
    buf = io.StringIO()
    try:
        with contextlib.redirect_stdout(buf):
            ret = fn(*args, **kwargs)
    except Exception as e:
        return {"error": e, "stdout": buf.getvalue(), "return": None}
    return {"error": None, "stdout": buf.getvalue(), "return": ret}

def _parse_freq_stdout(s):
    d = {}
    for line in s.splitlines():
        if "-->" in line:
            parts = line.split("-->")
            k = parts[0].strip()
            try:
                v = int(parts[1].strip())
            except:
                try:
                    v = int(parts[1].strip().strip())
                except:
                    continue
            d[k] = v
    return d

# Tests configured per problem (uses functions present in the notebook)
TEST_SUITE = {
    "Reverse String": {
        "fn": "reverse_string",
        "cases": [
            (("hello",), "olleh"),
            (("Python",), "nohtyP"),
            (("a",), "a"),
            (("",), ""),
        ],
        "type": "return"
    },
    "Check Armstrong Number": {
        "fn": "isArmstrong",
        "cases": [
            ((153,), True),
            ((9474,), True),
            ((123,), False),
            ((370,), True),
        ],
        "type": "return"
    },
    "Fibonacci Series": {
        "fn": "fibonacci_series",
        "cases": [
            ((7,), [0,1,1,2,3,5,8]),
            ((1,), [0,1]),
            ((0,), 0),
        ],
        "type": "return"
    },
    "Find Missing Number": {
        "fn": "findingmissing",
        "cases": [
            (([1,2,4,5],), 3),
            (([2,3,1,5],), 4),
            (([1],), 2),
        ],
        "type": "return"
    },
    "Count Frequency of Characters": {
        "fn": "freqcont",
        "cases": [
            (("hello",), {'h':1,'e':1,'l':2,'o':1}),
            (("aabbcc",), {'a':2,'b':2,'c':2}),
            (("",), {}),
        ],
        "type": "stdout_dict"  # parse printed output into dict
    },
    "Remove Duplicate Characters": {
        "fn": "removeduplicate",
        "cases": [
            (("banana",), "ban"),
            (("abcabc",), "abc"),
            (("aaaa",), "a"),
        ],
        "type": "return"
    },
    "Print Triangle/Star Patterns": {
        "fn": "print_triangle",
        "cases": [
            ((3,), "*\n**\n***\n"),
            ((5,), "*\n**\n***\n****\n*****\n"),
        ],
        "type": "stdout_str"
    },
    "Find Longest Word": {
        "fn": "longestword",
        "cases": [
            (("I love programming",), "programming"),
            (("AI is fun",), "fun"),
            (("",), ""),
        ],
        "type": "return"
    },
    "Check Anagrams": {
        "fn": "are_anagrams",
        "cases": [
            (("listen","silent"), True),
            (("hello","world"), False),
            (("aabb","baba"), True),
        ],
        "type": "return"
    },
    "Sum of Digits": {
        "fn": "sumdigits",
        "cases": [
            ((123,), 6),
            ((0,), 0),
            ((9999,), 36),
        ],
        "type": "return"
    },
}

report = {}
total_passed = 0
total_tests = 0

for prob, cfg in TEST_SUITE.items():
    fn_name = cfg["fn"]
    fn = _get_fn(fn_name)
    cases = cfg["cases"]
    passed = 0
    for args, expected in cases:
        total_tests += 1
        if not fn:
            # function missing
            continue
        result = _capture_stdout(fn, *args)
        if result["error"] is not None:
            continue
        if cfg["type"] == "return":
            out = result["return"]
            if out == expected:
                passed += 1
        elif cfg["type"] == "stdout_str":
            out = result["stdout"]
            # normalize newlines and trailing spaces
            if out == expected:
                passed += 1
        elif cfg["type"] == "stdout_dict":
            parsed = _parse_freq_stdout(result["stdout"])
            if parsed == expected:
                passed += 1
        else:
            # fallback compare return
            if result["return"] == expected:
                passed += 1
    total_passed += passed
    score = int(round(100 * (passed / len(cases)))) if cases else 0
    report[prob] = {"passed": passed, "total": len(cases), "score": score}

overall = int(round(100 * (total_passed / total_tests))) if total_tests else 0
report["__overall__"] = {"passed": total_passed, "total": total_tests, "score": overall}

# expose report variable and print summary
score_report = report
print(score_report)
````
