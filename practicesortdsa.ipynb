```python
#merge sort

def divcon(arrayInput):
    if len(arrayInput)<=1:
        return arrayInput
    mid=len(arrayInput)//2

    right=divcon(arrayInput[mid:])
    left=divcon(arrayInput[:mid])

    return merge(left, right)

def merge(left, right):
    leftPointer=0
    rightPointer=0
    result=[]

    while len(left)>leftPointer and len(right)>rightPointer:
        if left[leftPointer]<right[rightPointer]:
            result.append(left[leftPointer])
            leftPointer+=1

        else:
            result.append(right[rightPointer])
            rightPointer+=1
    result.extend(left[leftPointer:])
    result.extend(right[rightPointer:])
    return result

arrayInput=[453,534,546,85,95,5]
divcon(arrayInput)




    
```

```python
# selection sort
def selectionSort(arrayInput):
    if len(arrayInput)<=1:
        return arrayInput
    
    for i in range(len(arrayInput)):
        min_index=i
        for j in range(i, len(arrayInput)):
            if arrayInput[min_index]>arrayInput[j]:
                min_index=j
        arrayInput[i], arrayInput[min_index]=arrayInput[min_index], arrayInput[i]
    return arrayInput
arrayInput=[453,534,546,85,95,5]
selectionSort(arrayInput)


```

```python
# bubble sort

def bubbleSort(arrayInput):
    if len(arrayInput)<=1:
        return arrayInput
    for i in range(len(arrayInput)):
        swap=False
        for j in range(len(arrayInput)-1-i):
            if arrayInput[j]>arrayInput[j+1]:
                arrayInput[j], arrayInput[j+1]=arrayInput[j+1], arrayInput[j]
                swap=True
        if swap==False:
            return arrayInput
    return arrayInput

arrayInput=[453,534,546,85,1, 95,5]
bubbleSort(arrayInput)

```

```python
# quick sort
import numpy as np
def quickSort(arrayInput):
    if len(arrayInput)<=1:
        return arrayInput
    left=[]
    right=[]
    pivot_index=np.random.randint(0, len(arrayInput)-1)
    pivot=arrayInput[pivot_index]
    for i in range(len(arrayInput)):
        if i==pivot_index:
            continue
        if arrayInput[i]<pivot:
            left.append(arrayInput[i])
        else:
            right.append(arrayInput[i])
    return quickSort(left) + [pivot] + quickSort(right)

arrayInput=[453,534,546,85,95,5, 453,54]
quickSort(arrayInput)
```
