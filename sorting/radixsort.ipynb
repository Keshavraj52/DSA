## ðŸ”‘ Key Idea of Radix Sort
Radix sort works by sorting numbers digit by digit, starting from the **least significant digit (LSD)** to the **most significant digit (MSD)** (or vice versa, depending on the variant).  
It uses a **stable sorting algorithm** (commonly counting sort) at each digit level.

Think of it as organizing exam papers first by the **units place**, then by the **tens place**, then by the **hundreds place**, until everything is sorted.

## âš™ï¸ Python Implementation

```python
def counting_sort(arr, exp):
    n = len(arr)
    output = [0] * n
    count = [0] * 10  # digits 0-9

    # Count occurrences of each digit
    for i in range(n):
        index = arr[i] // exp
        count[index % 10] += 1

    # Update count[i] to contain actual position
    for i in range(1, 10):
        count[i] += count[i - 1]

    # Build output array (stable sort)
    i = n - 1
    while i >= 0:
        index = arr[i] // exp
        output[count[index % 10] - 1] = arr[i]
        count[index % 10] -= 1
        i -= 1

    # Copy output back to arr
    for i in range(n):
        arr[i] = output[i]

def radix_sort(arr):
    # Find maximum number to know number of digits
    max_num = max(arr)

    # Apply counting sort for each digit
    exp = 1
    while max_num // exp > 0:
        counting_sort(arr, exp)
        exp *= 10

    return arr

# Example usage
data = [170, 45, 75, 90, 802, 24, 2, 66]
print(radix_sort(data))  # [2, 24, 45, 66, 75, 90, 170, 802]
```
