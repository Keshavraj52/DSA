## ðŸ”‘ Key Idea of Bucket Sort
Bucket sort distributes elements into a number of â€œbuckets,â€ sorts each bucket individually (often using insertion sort or another simple algorithm), and then concatenates the buckets to form the final sorted array.

It works best when input is **uniformly distributed** over a range.

## âš™ï¸ Python Implementation

```python
def insertion_sort(bucket):
    for i in range(1, len(bucket)):
        key = bucket[i]
        j = i - 1
        while j >= 0 and bucket[j] > key:
            bucket[j + 1] = bucket[j]
            j -= 1
        bucket[j + 1] = key
    return bucket

def bucket_sort(arr):
    n = len(arr)
    if n == 0:
        return arr

    # Create n empty buckets
    buckets = [[] for _ in range(n)]

    # Insert elements into buckets
    for num in arr:
        index = int(num * n)  # scale value to bucket index
        buckets[index].append(num)

    # Sort each bucket and concatenate
    sorted_arr = []
    for bucket in buckets:
        if bucket:
            insertion_sort(bucket)
            sorted_arr.extend(bucket)

    return sorted_arr

# Example usage
data = [0.78, 0.17, 0.39, 0.26, 0.72, 0.94, 0.21, 0.12, 0.23, 0.68]
print(bucket_sort(data))  
# Output: [0.12, 0.17, 0.21, 0.23, 0.26, 0.39, 0.68, 0.72, 0.78, 0.94]
```

## ðŸ“Œ Characteristics
- **Time complexity:**  
  - Average: \(O(n + k)\) (if distribution is uniform and buckets are balanced).  
  - Worst case: \(O(n^2)\) (if all elements fall into one bucket).
- **Stable:** Depends on the sorting algorithm used inside buckets.
- **Space complexity:** \(O(n + k)\) for buckets.
- **Best use case:** Sorting uniformly distributed floating-point numbers.
