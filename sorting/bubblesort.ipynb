
Concise explanation and implementations of Bubble Sort.

Bubble Sort summary:
- Repeatedly steps through the list, compares adjacent pairs and swaps them if they are in the wrong order.
- After each full pass the largest (for ascending) unsorted element "bubbles" to its final position.
- Stable: equal elements keep their original relative order.
- In-place: uses O(1) extra space (ignoring input/output).
- Time complexity:
    - Worst-case: O(n^2) (reverse-sorted)
    - Average-case: O(n^2)
    - Best-case: O(n) for optimized version when input is already sorted
- Use cases: educational purposes, tiny arrays, or when stability and simplicity are needed. For large arrays, prefer O(n log n) sorts.

```python
def bubbleSort(arr):
    n=len(arr)
    for i in range(n-1):##[1,5,12]
        swap=0

        for j in range(n-1-i):
            if arr[j]>arr[j+1]:
                arr[j], arr[j+1]=arr[j+1], arr[j]
                swap=1
        if swap==0:
            break
    return arr


arr=[45,65,23,87,78]
bubbleSort(arr)
        
```
