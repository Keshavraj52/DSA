## ğŸ”‘ Key Idea of Heap Sort
Heap sort uses a **binary heap** (usually a max-heap) to repeatedly extract the largest element and place it at the end of the array.  
Steps:
1. Build a max-heap from the array.
2. Swap the root (largest element) with the last element.
3. Reduce heap size by 1 and heapify again.
4. Repeat until the array is sorted.

---

## ğŸ“Š Step-by-Step Example
Letâ€™s sort `[4, 10, 3, 5, 1]`.

### Step 1: Build Max-Heap
- Original array: `[4, 10, 3, 5, 1]`
- After heapify: `[10, 5, 3, 4, 1]` (10 is root, largest element)

### Step 2: Extract Max (10)
- Swap root with last element â†’ `[1, 5, 3, 4, 10]`
- Heapify first 4 elements â†’ `[5, 4, 3, 1, 10]`

### Step 3: Extract Max (5)
- Swap root with last element in heap â†’ `[1, 4, 3, 5, 10]`
- Heapify first 3 elements â†’ `[4, 1, 3, 5, 10]`

### Step 4: Extract Max (4)
- Swap root with last element in heap â†’ `[3, 1, 4, 5, 10]`
- Heapify first 2 elements â†’ `[3, 1, 4, 5, 10]`

### Step 5: Extract Max (3)
- Swap root with last element in heap â†’ `[1, 3, 4, 5, 10]`
- Heapify first 1 element â†’ `[1, 3, 4, 5, 10]`

âœ… Final sorted array: `[1, 3, 4, 5, 10]`

## âš™ï¸ Python Implementation

```python
def heapify(arr, n, i):
    largest = i
    left = 2 * i + 1
    right = 2 * i + 2

    # Check if left child exists and is greater
    if left < n and arr[left] > arr[largest]:
        largest = left

    # Check if right child exists and is greater
    if right < n and arr[right] > arr[largest]:
        largest = right

    # If largest is not root, swap and continue heapifying
    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        heapify(arr, n, largest)

def heap_sort(arr):
    n = len(arr)

    # Build max heap
    for i in range(n // 2 - 1, -1, -1):
        heapify(arr, n, i)

    # Extract elements one by one
    for i in range(n - 1, 0, -1):
        arr[0], arr[i] = arr[i], arr[0]  # swap
        heapify(arr, i, 0)

    return arr

# Example usage
data = [4, 10, 3, 5, 1]
print(heap_sort(data))  # [1, 3, 4, 5, 10]
```

## ğŸ“Œ Characteristics
- **Time complexity:**  
  - Building heap: \(O(n)\)  
  - Extracting max repeatedly: \(O(n \log n)\)  
  - Overall: \(O(n \log n)\) worst, average, and best case.
- **Not stable:** Equal elements may not preserve order.
- **In-place:** Requires only constant extra space.
