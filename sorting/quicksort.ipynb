## ðŸ”Ž How Quick Sort Works
1. **Choose a Pivot**  
   - Select an element from the array (first, last, random, or median).  
   - Example: In `[10, 7, 8, 9, 1, 5]`, choose `last element = 5` as pivot.

2. **Partition the Array**  
   - Rearrange elements so that all values **smaller than pivot** go to its left, and all values **greater than pivot** go to its right.  
   - After partitioning: `[1, 5, 8, 9, 10, 7]` â†’ pivot `5` is in its correct position.

3. **Recursive Calls**  
   - Apply the same process to the left subarray `[1]` and right subarray `[8, 9, 10, 7]`.  
   - Continue until subarrays have one or zero elements (base case).

## ðŸ§© Example Walkthrough
Letâ€™s sort `[10, 7, 8, 9, 1, 5]` step by step:

- **Step 1:** Pivot = `5`  
  Partition â†’ `[1, 5, 10, 7, 8, 9]`  

- **Step 2:** Left subarray `[1]` (already sorted).  
  Right subarray `[10, 7, 8, 9]` â†’ choose pivot `9`.  
  Partition â†’ `[7, 8, 9, 10]`.

- **Step 3:** Left subarray `[7, 8]` â†’ pivot `8`.  
  Partition â†’ `[7, 8]` (sorted).  
  Right subarray `[10]` (sorted).

âœ… Final sorted array: `[1, 5, 7, 8, 9, 10]`.

---

## âš¡ Key Points
- **Best/Average Case:** \(O(n \log n)\)  
- **Worst Case:** \(O(n^2)\) (when pivot is poorly chosen, e.g., always smallest/largest).  
- **Space Complexity:** \(O(\log n)\) due to recursion stack.  
- **Efficiency:** Faster than bubble/selection sort for large datasets.

```python
def quicksort(arr):
    if len(arr)<=1:
        return arr
    pivot=arr[-1]
    left=[]
    right=[]
    for i in range(len(arr)-1):
        if pivot>arr[i]:
            left.append(arr[i])
        else:
            right.append(arr[i])
    return quicksort(left)+[pivot]+quicksort(right)


arr=[45,18,7,54,12,6,2,1]
quicksort(arr)
```

```python
import numpy as np
def quicksort(arr):
    if len(arr)<=1:
        return arr
    pivot_index=np.random.randint(0,len(arr)-1)
    pivot=arr[pivot_index]
    left=[]
    right=[]
    for i in range(len(arr)):
        if i==pivot_index:
            continue
        if pivot>arr[i]:
            left.append(arr[i])
        else:
            right.append(arr[i])
    return quicksort(left)+[pivot]+quicksort(right)


arr=[45,18,7,54,12,6,2,1]
quicksort(arr)
```
