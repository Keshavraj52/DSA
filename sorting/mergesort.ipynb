
## ðŸ”¹ How Merge Sort Works
Merge Sort is a **divide-and-conquer algorithm**:
1. **Divide**: Split the array into two halves until each subarray has only one element.
2. **Conquer**: Recursively sort each half.
3. **Combine (Merge)**: Merge the sorted halves into a single sorted array.

---

## ðŸ”¹ Example Array
Letâ€™s take this array:  
```
[38, 27, 43, 3, 9, 82, 10]
```

---

## ðŸ”¹ Step-by-Step Process

### 1. Divide
Split into halves:
```
[38, 27, 43]   and   [3, 9, 82, 10]
```

Further split:
```
[38, 27] and [43]      |      [3, 9] and [82, 10]
```

Split again:
```
[38] and [27]          |      [3] and [9]      |      [82] and [10]
```

---

### 2. Conquer (Sort each small array)
Single-element arrays are already sorted:
```
[38], [27], [43], [3], [9], [82], [10]
```

---

### 3. Merge
Now merge step by step:

- Merge `[38]` and `[27]` â†’ `[27, 38]`
- Merge `[3]` and `[9]` â†’ `[3, 9]`
- Merge `[82]` and `[10]` â†’ `[10, 82]`

Now we have:
```
[27, 38], [43], [3, 9], [10, 82]
```

---

### 4. Merge Again
- Merge `[27, 38]` and `[43]` â†’ `[27, 38, 43]`
- Merge `[3, 9]` and `[10, 82]` â†’ `[3, 9, 10, 82]`

Now we have:
```
[27, 38, 43] and [3, 9, 10, 82]
```

---

### 5. Final Merge
Merge `[27, 38, 43]` and `[3, 9, 10, 82]` â†’  
```
[3, 9, 10, 27, 38, 43, 82]
```

---

## âœ… Final Sorted Array
```
[3, 9, 10, 27, 38, 43, 82]
```

---

## ðŸ”¹ Key Points
- **Time Complexity**: \(O(n \log n)\)  
- **Space Complexity**: \(O(n)\) (needs extra space for merging)  
- **Stable Sort**: Yes (preserves order of equal elements)

---


````python
def divCon(arr):
    if len(arr)<=1:
        return arr
    mid=len(arr)//2
    left=divCon(arr[:mid])
    right=divCon(arr[mid:])
    return merge(left, right)

def merge(left,right):
    i=0
    j=0
    result=[]
    while len(left)>i and len(right)>j:
        if left[i]< right[j]:
            result.append(left[i])
            i+=1
        else:
            result.append(right[j])
            j+=1
    result.extend(left[i:])
    result.extend(right[j:])
    return result

array=[45,54,2,65,8,32,44,45]
divCon(array)
    

````
