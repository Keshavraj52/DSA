
### Insertion sort step-by-step on a small array

We’ll sort the array `[5, 3, 4, 1, 2]` in ascending order. The left side is the “sorted” portion that grows each step; we insert the next element into its correct position by shifting larger elements to the right.

#### Pass 1: Insert 3 into [5]
- **Key:** 3  
- **Compare:** 3 < 5 → shift 5 right  
- **After shifts:** `[5, 5, 4, 1, 2]`  
- **Insert key:** place 3 at index 0  
- **State:** `[3, 5, 4, 1, 2]`

#### Pass 2: Insert 4 into [3, 5]
- **Key:** 4  
- **Compare:** 4 < 5 → shift 5 right; 4 ≥ 3 → stop  
- **After shifts:** `[3, 5, 5, 1, 2]`  
- **Insert key:** place 4 at index 1  
- **State:** `[3, 4, 5, 1, 2]`

#### Pass 3: Insert 1 into [3, 4, 5]
- **Key:** 1  
- **Compare:** 1 < 5 → shift 5; 1 < 4 → shift 4; 1 < 3 → shift 3  
- **After shifts:** `[3, 3, 4, 5, 2]`  
- **Insert key:** place 1 at index 0  
- **State:** `[1, 3, 4, 5, 2]`

#### Pass 4: Insert 2 into [1, 3, 4, 5]
- **Key:** 2  
- **Compare:** 2 < 5 → shift 5; 2 < 4 → shift 4; 2 < 3 → shift 3; 2 ≥ 1 → stop  
- **After shifts:** `[1, 1, 3, 4, 5]`  
- **Insert key:** place 2 at index 1  
- **Final sorted:** `[1, 2, 3, 4, 5]`


---

### Notes on complexity and properties
- **Best case:** already sorted, \(\mathcal{O}(n)\) comparisons and minimal shifts.
- **Average/worst case:** \(\mathcal{O}(n^2)\).
- **Stable and in-place:** preserves equal-order and uses constant extra space. 



```python
def insertionSort(arr):
    if len(arr)<=1:
        return arr
    for i in range(1,len(arr)):
        key=arr[i]
        j=i-1
        while j>=0 and arr[j]>key:
            arr[j+1]=arr[j]
            j-=1
        arr[j+1]=key
    return arr

arr=[45,54,7,10,12,18]
insertionSort(arr)
```
