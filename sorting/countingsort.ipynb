## ğŸ”‘ Key Idea of Counting Sort
Counting sort is a **non-comparison sorting algorithm**.  
It works by:
1. Counting how many times each element appears.
2. Using those counts to determine the position of each element in the output array.
3. Placing elements directly into their sorted position.

Itâ€™s efficient when the range of input values (max âˆ’ min) is not significantly larger than the number of elements.

## ğŸ“Š Step-by-Step Example
Letâ€™s sort `[4, 2, 2, 8, 3, 3, 1]`.

### Step 1: Find range
- Minimum = 1, Maximum = 8  
- Range = 8 âˆ’ 1 + 1 = 8

### Step 2: Initialize count array
- Count array of size 8 (all zeros):  
  `[0, 0, 0, 0, 0, 0, 0, 0]`

### Step 3: Count occurrences
- For each element in input:  
  - 4 â†’ count[3]++  
  - 2 â†’ count[1]++  
  - 2 â†’ count[1]++  
  - 8 â†’ count[7]++  
  - 3 â†’ count[2]++  
  - 3 â†’ count[2]++

- Count array becomes:  
  `[1, 2, 2, 1, 0, 0, 0, 1]`

### Step 4: Cumulative count
- Transform count array to cumulative positions:  
  `[1, 3, 5, 6, 6, 6, 6, 7]`  
  (Each entry tells how many elements â‰¤ that index.)

### Step 5: Place elements into output
- Traverse input backwards (to maintain stability):  
  - 1 â†’ position 0 â†’ `[1, _, _, _, _, _, _]`  
  - 3 â†’ position 4 â†’ `[1, _, _, _, 3, _, _]`  
  - 3 â†’ position 3 â†’ `[1, _, _, 3, 3, _, _]`

## âš™ï¸ Python Implementation

```python
def counting_sort(arr):
    # Find the maximum and minimum values
    max_val = max(arr)
    min_val = min(arr)
    range_val = max_val - min_val + 1

    # Initialize count array
    count = [0] * range_val

    # Count occurrences
    for num in arr:
        count[num - min_val] += 1

    # Cumulative count
    for i in range(1, range_val):
        count[i] += count[i - 1]

    # Output array
    output = [0] * len(arr)

    # Place elements (stable sort, traverse backwards)
    for num in reversed(arr):
        count[num - min_val] -= 1
        output[count[num - min_val]] = num

    return output

# Example usage
data = [4, 2, 2, 8, 3, 3, 1]
print(counting_sort(data))  # [1, 2, 2, 3, 3, 4, 8]
```

## ğŸ“Œ Characteristics
- **Time complexity:** \(O(n + k)\), where \(n\) = number of elements, \(k\) = range of values.
- **Stable:** Yes (if implemented with backward traversal).
- **Not comparison-based:** Works only for integers or categorical data with a limited range.
- **Space complexity:** \(O(n + k)\) due to count + output arrays.
